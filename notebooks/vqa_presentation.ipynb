## 1. Giới thiệu
**Visual Question Answering (VQA)** là bài toán kết hợp thị giác máy tính và xử lý ngôn ngữ tự nhiên: mô hình nhận **ảnh** và **câu hỏi**, sau đó sinh ra **câu trả lời**.

**Tại sao VQA là bài toán đa phương thức (multimodal)?**
- Ảnh và văn bản có bản chất dữ liệu khác nhau.
- Mô hình phải học cách **kết hợp** thông tin thị giác và ngôn ngữ để suy luận câu trả lời.

**Thách thức của VQA tiếng Việt**
- Tiếng Việt có đặc điểm tách từ phức tạp (word segmentation).
- Tài nguyên dữ liệu và mô hình pretrained hạn chế hơn so với tiếng Anh.

**Mục tiêu dự án**
- Xây dựng pipeline VQA hoàn chỉnh, rõ ràng, có thể tái lập.
- Thử nghiệm nhiều biến thể mô hình (CNN/ResNet, có/không attention).
- Đánh giá định lượng và định tính, phục vụ demo.

**Sơ đồ khái niệm**
```
Image + Question  -->  VQA Model  -->  Answer
```

## 2. Mô hình hóa bài toán
Cho ảnh $I$ và câu hỏi $Q = (q_1, q_2, ..., q_n)$, mô hình sinh câu trả lời $A = (a_1, a_2, ..., a_m)$.

Mục tiêu là tối đa hóa xác suất có điều kiện:
$$P(A | I, Q) = prod_{t=1}^{m} P(a_t | a_{<t}, I, Q)$$

Vì câu trả lời là một chuỗi biến độ dài, bài toán được xử lý như **sequence generation** (seq2seq), thay vì classification.

## 4. Tiền xử lý dữ liệu
Các bước tiền xử lý:
1. Chuẩn hóa (lowercase, bỏ khoảng trắng dư)
2. Tokenize tiếng Việt bằng `underthesea`
3. Xây dựng từ vựng
4. Thêm token đặc biệt: `<PAD>`, `<START>`, `<END>`, `<UNK>`
5. Padding về độ dài cố định

**Ví dụ**
- Câu gốc: "có bao nhiêu người đàn ông"

**Tại sao cần padding?**
- Batch training yêu cầu mọi chuỗi có cùng độ dài.
- `<PAD>` được bỏ qua khi tính loss.

## 7. Question Encoder
Pipeline xử lý câu hỏi:
Tokenized question → Embedding → LSTM encoder

LSTM tạo ra hidden states theo từng bước thời gian.
Hidden state cuối cùng đại diện cho toàn câu hỏi.

**Sơ đồ LSTM (ASCII)**
```
q1 -> [LSTM] -> h1
q2 -> [LSTM] -> h2
...
qn -> [LSTM] -> hn (final representation)
```

## 8. Multimodal Fusion
Kết hợp đặc trưng ảnh và câu hỏi bằng cách **concatenate**:
$$f = 	anh(W [v_{image}; v_{question}])$$

**Tại sao cần fusion?**
- Cho phép mô hình tạo một không gian chung giữa thị giác và ngôn ngữ.
- Cải thiện khả năng suy luận đa phương thức.

## 9. Attention Mechanism (Bahdanau)
**Mục tiêu**: tập trung vào các vùng ảnh liên quan đến câu hỏi.

**Công thức**
$$score = v^T 	anh(W_1 * image + W_2 * hidden)$$

**Các bước**
1. Tính score cho từng vùng ảnh
2. Softmax để tạo attention weights
3. Tính context vector bằng weighted sum

**Sơ đồ**
```
Regions: r1 r2 r3 ... r196
Weights: w1 w2 w3 ... w196
Context = sum(wi * ri)
```

## 10. Decoder
Decoder LSTM sinh từng token một cách tự hồi quy.

Tại thời điểm t:
- Input: token trước đó + context vector
- Output: phân phối xác suất trên vocab

**Teacher Forcing**
- Trong training, đôi khi dùng token thật thay vì token dự đoán.
- Giúp ổn định huấn luyện và tăng tốc hội tụ.

## 11. Beam Search
Greedy decoding chọn token có xác suất lớn nhất ở mỗi bước, nhưng có thể bỏ lỡ chuỗi tối ưu.

**Beam search** (beam width = 3):
- Giữ 3 chuỗi ứng viên tốt nhất ở mỗi bước
- Chọn chuỗi có xác suất cao nhất cuối cùng

**Ví dụ (rút gọn)**
1. Step 1: chọn 3 token tốt nhất
2. Step 2: mở rộng 3 chuỗi × top-3 token
3. Giữ lại top-3 chuỗi có log-prob cao nhất

## 13. Training
**Tham số chính**

**Kỹ thuật hỗ trợ**
- Teacher forcing schedule (giảm dần theo epoch)
- AMP mixed precision để tiết kiệm VRAM và tăng tốc
- Gradient clipping để ổn định huấn luyện
- Checkpointing để resume khi crash
